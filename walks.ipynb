```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
import glob
files = glob.glob("av_files/*.av")
```

```python
import os
path = 'av_files/DATA_short'
files = os.listdir(path)
```

```python
f = open(path + '/' + files[0])
r, r2=[], []
f.readline()
for line in f:
    r.append(float(line.split()[0]))
    r2.append(float(line.split()[1]))
        
arr = list(zip(r, r2))
df = pd.DataFrame(arr, columns = ['r', 'r2'])
```

```python
plt.plot(df.r, df.r2)
```

```python
slopes = []
samples = []
sample = '0'
ids = []
idxs = []
idx = '020'

names = ['0']
i = 0
for file in files:
    f = open(path + '/' + file)
    r, r2=[], []
    f.readline()
    for line in f:
        r.append(float(line.split()[0]))
        r2.append(float(line.split()[1]))
    slope, intercept, _, _, stderr = stats.linregress(np.array(r), np.array(r2))

    if sample.lower() == file[0].lower():
        slopes.append(slope)
        ids.append(file[1:4])
    else:
        sample = file[0]
        samples.append(slopes)
        slopes = [] 
        slopes.append(slope)
        idx = file[1:4]
        idxs.append(ids)
        ids = []
        ids.append(idx)
        names.append(file[0])
    print('slope for r ', slope, file[0], file[1:4])
    plt.plot(r, r2)
    plt.title('All in one')
```

```python
for sample, name in zip(samples, names):
    plt.plot(sample, label = name)
#plt.legend()
plt.title('Slopes for each sample')
plt.show()
```

```python
for sample, name in zip(samples, names):
    if name == '0':
        plt.plot(sample[1:], label = name)
    else:
        if np.array(sample).min() > 0:
            plt.plot(sample, label = name)
#plt.legend()
plt.yscale('log')
plt.title('Slopes for each sample (without one first value for A) ')
plt.xlabel('i')
plt.ylabel('alpha = D/6')
```

```python
f_out = open("slopes.txt", "w")
f_out.write("slope sample slice mean std take_throw\n")
for sample, name, idx in zip(samples, names, idxs):
    std = np.array(sample).std()
    mean = np.array(sample).mean()
    minima = np.array(sample).min()
    if minima < 0:
        take = 0
    else:
        take = 1
    for slope, id_ in zip(sample, idx):
        f_out.write("%1.8f %s %s %f %f %d\n" % (slope, name, id_, mean, std, take))
f_out.close()
```
