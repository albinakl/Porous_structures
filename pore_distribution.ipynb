```python
data = {
  "tot_por": [],
  "av_por_size": [],
  "av_perim_rel": [],
  "av_rad_rel": [],
  "sample": [],
  "slice": []
}
for file in files:
    f = open(path + '/' + file)
    print(file)
    r, s, p=[], [], []
    f.readline()
    for line in f:
        r.append(float(line.split()[0]))
        s.append(float(line.split()[1]))
        p.append(float(line.split()[2]))
    f.close()
    r_ = np.array(r)
    s_ = np.array(s)
    p_ = np.array(p)
    
    data['tot_por'].append(s_.sum()/193**3)
    data['av_por_size'].append(r_.mean())
    data['av_perim_rel'].append(((p_*p_)/s_).mean())
    data['av_rad_rel'].append(((r_*r_)/s_).mean())
    data['sample'].append(file[0])
    data['slice'].append(int(file[1:4]))
```

```python
dataframe = pd.DataFrame.from_dict(data)
```

## Form Y dataframe

```python
y_col = pd.read_csv("slopes.txt", delimiter = " ")
y_col['diffusion'] = y_col['slope'] * 6
```

```python
y_col
```

### merge this dataframes (drop missing and wrong rows)

```python
y_col = y_col[y_col['take_throw'] == 1].copy()
```

```python
y_col.shape[0], dataframe.shape[0]
```

```python
y_col['final_des'] = np.zeros(y_col.shape[0])
dataframe['final_des'] = np.zeros(dataframe.shape[0])
```

```python
for i in range(y_col.shape[0]):
    sample = y_col['sample'].iloc[i]
    slice_ = y_col['slice'].iloc[i]
    
    if dataframe[(dataframe['sample']==sample) & (dataframe['slice']==slice_)].shape[0] == 1:
        y_col['final_des'].iloc[i] = 1
        dataframe.loc[(dataframe['sample']==sample) & (dataframe['slice']==slice_), 'final_des'] = 1
```

```python
X_full = dataframe[dataframe['final_des'] == 1].copy()
y_full = y_col[y_col['final_des'] == 1]['diffusion'].copy()
```

```python
X_full.reset_index(drop=True, inplace=True)
X_full
```

```python
y_full.reset_index(drop=True, inplace=True)
y_full
```

## Feature selection

unificate sample column

```python
arr = X_full['sample'].to_list()
```

```python
new_arr = []
for el in arr:
    new_arr.append(el.lower())
```

```python
X_full['sample'] = pd.Series(new_arr)
```

```python
X_full
```

## Train Test split

```python
n_samples = len(X_full['sample'].unique())
n_test_samples = int(n_samples * 0.3)
n_train_samples = n_samples - n_test_samples
n_train_samples, n_test_samples
```

```python
train_samples_names = list(X_full['sample'].unique())[:n_train_samples]
test_samples_names = list(X_full['sample'].unique())[-n_test_samples:]
train_samples_names, test_samples_names
```

```python
X_train = X_full[ X_full['sample'].isin(train_samples_names) ].copy()
X_test = X_full[ X_full['sample'].isin(test_samples_names) ].copy()
```

```python
y_train = y_full.iloc[:X_train.shape[0]].copy()
y_test = y_full.iloc[-X_test.shape[0]:].copy()
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape 
```

```python
X_train.drop(['final_des', 'slice', 'sample'], axis=1, inplace=True)
X_test.drop(['final_des', 'slice', 'sample'], axis=1, inplace=True)
X_train
```

## Train the model

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.preprocessing import StandardScaler as SS
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split
```

```python
train_X = np.array(X_train)
test_X = np.array(X_test)
train_y = np.array(y_train)
test_y = np.array(y_test)
```

```python
scaler = SS()
scaler.fit(train_X)
train_X = scaler.transform(train_X)
test_X = scaler.transform(test_X)
```

```python
gs_boost = GridSearchCV(GradientBoostingRegressor(), 
            param_grid = {'n_estimators': [500, 300],
                          'max_depth': [2, 3, 4],
                          'min_samples_split': [3, 5, 7],
                          'learning_rate': [0.01, 0.02, 0.05, 0.1, 0.2],
                          'loss': ['ls']},
            cv = KFold(n_splits = 3, 
                       shuffle = True, 
                       random_state = 42), 
            n_jobs = -1, 
            verbose = 1)

gs_boost.fit(train_X, train_y)
preds = gs_boost.predict(test_X)
```

```python
preds = gs_boost.predict(test_X)
```

```python
gs_boost.best_score_, gs_boost.best_estimator_.score(test_X, test_y)
```

```python
mse(test_y, preds)
```

```python
gs_boost.best_params_
```

```python
imp = np.argmax(gs_boost.best_estimator_.feature_importances_)
X_train.columns[imp]
```

```python
test_score = np.zeros((gs_boost.best_params_['n_estimators'],), dtype=np.float64)
for i, pred in enumerate(gs_boost.best_estimator_.staged_predict(test_X)):
    test_score[i] = gs_boost.best_estimator_.loss_(test_y, pred)

fig = plt.figure(figsize=(6, 6))
plt.subplot(1, 1, 1)
plt.title('Deviance')
plt.plot(np.arange(gs_boost.best_params_['n_estimators']) + 1, gs_boost.best_estimator_.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(gs_boost.best_params_['n_estimators']) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')
fig.tight_layout()
plt.show()
```
